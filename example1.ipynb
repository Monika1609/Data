```python
filepath=("/content/sample_data/example.xlsx")
df=pd.read_excel("/content/sample_data/example.xlsx")
df
```

```python
x=df[["points","assits","rebounds"]]
y=df["rating"]
```

```python
from sklearn.linear_model import LinearRegression
#from sklearn.model_selection import train_test_split
model=LinearRegression()
model.fit(x,y)
```

```python
y_pred=model.predict(x)
```

```python
from sklearn.metrics import r2_score
r2_score(y,y_pred)
```
