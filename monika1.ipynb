```python
#mersing two different dataframes
res=pd.get_dummies(df1["sex"],dtype=int)
new_df=df1.join(res)
new_df.drop("sex",axis=1,inplace=True)
new_df.head()
```

```python
res.head()
```

```python
#replace smokers with numerical value
new_df["smoker"]=new_df["smoker"].map({"yes":1,"no":0})
new_df.head()
```

```python
#replace region with numerical values
new_df["region"].value_counts()
new_df["region"]=new_df["region"].map({"southeast":1,"southwest":2,"northwest":3,"northeast":4})
new_df.head()
```

```python
x=np.array(new_df["bmi"]).reshape(-1,1)
y=np.array(new_df["charges"]).reshape(-1,1)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.25,random_state=42)
```

```python
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
x_train.shape
```

```python
#training score
model.score(x_train,y_train)
```

```python
#testing score
model.score(x_test,y_test)
```

```python
#model accuracy score

from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```
